```python
#import module 'cv2' as cv
import cv2 as cv
#import module 'numpy' as np
import numpy as np
#import the pyplot module from the matplotlib library and assigns it the alias plt
from matplotlib import pyplot as plt

#takes the image as an input
def hist_values(image):
   
    #flatten the image array in row-major (C-style) order
    flat_image = image.flatten('C')
    
    #find unique intensity values and their counts using np.unique
    intensity, count = np.unique(flat_image, return_counts=True)
    
    #return the intensity values and counts
    return intensity, count

#read the image in BGR format (default in opencv)
bgr_test_pattern = cv.imread('test_pattern.tif', cv.IMREAD_COLOR)

#convert the image from BGR to RGB format
rgb_test_pattern = cv.cvtColor(bgr_test_pattern, cv.COLOR_BGR2RGB) 

#generate random samples from the Gaussian distribution
gaussian_noise = np.random.normal(0, 100, rgb_test_pattern.shape)

#generate random samples from the Uniform distribution
uniform_noise = np.random.uniform(0, 0.1, rgb_test_pattern.shape)

#generate random samples from the Rayleigh distribution
rayleigh_noise = np.random.rayleigh(0.5, rgb_test_pattern.shape)

#generate random samples from the exponential distribution
exponential_noise = np.random.exponential(1, rgb_test_pattern.shape)

#normalize the gaussian_noise array and convert it to uint8
normalized_gaussian_noise = np.uint8(cv.normalize(gaussian_noise, None, 0, 200, cv.NORM_MINMAX, dtype=cv.CV_32F))

#normalize the uniform_noise array and convert it to uint8
normalized_uniform_noise = np.uint8(cv.normalize(uniform_noise, None, 0, 50, cv.NORM_MINMAX, dtype=cv.CV_32F))

#normalize the rayleigh_noise array and convert it to uint8
normalized_rayleigh_noise = np.uint8(cv.normalize(rayleigh_noise, None, 0, 50, cv.NORM_MINMAX, dtype=cv.CV_32F))

#normalize the exponential_noise array and convert it to uint8
normalized_exponential_noise = np.uint8(cv.normalize(exponential_noise, None, 0, 100, cv.NORM_MINMAX, dtype=cv.CV_32F))

#overlay normalized gaussian noise over the image
gaussian_noisy_image = cv.add(bgr_test_pattern, normalized_gaussian_noise)

#overlay normalized uniform noise over the image
uniform_noisy_image = cv.add(bgr_test_pattern, normalized_uniform_noise)

#overlay normalized rayleigh noise over the image
rayleigh_noisy_image = cv.add(bgr_test_pattern, normalized_rayleigh_noise)

#overlay normalized exponential noise over the image
exponential_noisy_image = cv.add(bgr_test_pattern, normalized_exponential_noise)

#call the hist_values function to compute intensity values and counts
def plot_image (image, noise_label, noise, noisy_image, row):
    intensity, count = hist_values(noisy_image)

    fig = plt.figure(dpi=900)
    
    fig.add_subplot(row, 4, 1)
    plt.imshow(image.astype(np.uint8))
    plt.axis("off")
    plt.title("Original", fontsize = 10)

    fig.add_subplot(row, 4, 2)
    plt.axis("off")
    plt.imshow(noise)
    plt.title(noise_label, fontsize = 10)

    fig.add_subplot(row, 4, 3)
    plt.axis("off")
    plt.imshow(noisy_image.astype(np.uint8))
    plt.title("Noisy Image", fontsize = 10)
    
    
def plot_histogram (images, image_labels):
    #create subplots with 1 row and 4 columns
    fig, axs = plt.subplots(1, 4, figsize=(10, 3))

    #iterate over the images and labels
    for i in range(4):
        #compute the histogram
        hist, bins = np.histogram(images[i].flatten(), bins=range(256))

        #plot the histogram using bar plot
        axs[i].bar(bins[:-1], hist)

        #set the label for each subplot
        axs[i].set_title(labels[i])

        #set x and y axis labels
        axs[i].set_xlabel('Intensity')
        axs[i].set_ylabel('Count')

    #adjust the spacing between subplots
    plt.tight_layout()

    #display the plot
    plt.show()



images = [gaussian_noisy_image, uniform_noisy_image, rayleigh_noisy_image, exponential_noisy_image]
labels = ['Gaussian', 'Uniform', 'Rayleigh', 'Exponential']

#cv.calcHist(images, [0, 1, 2], None, [256], )

# Call the function to plot histograms
colors = 'b', 'g', 'r'
plot_histogram(images, labels)

plot_image(rgb_test_pattern, 'Gaussian Noise', normalized_gaussian_noise, gaussian_noisy_image, 1)
plot_image(rgb_test_pattern, 'Uniform Noise', normalized_uniform_noise, uniform_noisy_image, 2)
plot_image(rgb_test_pattern, 'Rayliegh Noise', normalized_rayleigh_noise, rayleigh_noisy_image, 3)
plot_image(rgb_test_pattern, 'Exponential Noise', normalized_exponential_noise, exponential_noisy_image, 4)
```

```python
import numpy as np
import cv2 as cv

#generate random integers in the range [0, 256) of 7x7 size
random_array = np.random.randint(0, 256, (7, 7), dtype = np.uint8)
print("First random 7x7 integers:\n", random_array, "\n")

#create a 2D array filled with ones 0f 3x3 size
filter = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])

#np.ones((3, 3), dtype=np.float32)
print('First Filter: \n', filter, '\n')

result = cv.filter2D(random_array, ddepth=-1, kernel=filter, anchor=(-1, -1), borderType=cv.BORDER_ISOLATED)
print("First Convolved result: \n", result, '\n')

random_array = np.random.randint(0, 256, (7, 7), dtype = np.uint8)
print("Second random 7x7 integers:\n", random_array, "\n")

filter = np.array([[0, 1, 0], [1, 0, 1], [0, 1, 0]])
print('Second Filter: \n', filter, '\n')

result = cv.filter2D(random_array, ddepth=-1, kernel=filter, anchor=(-1, -1), borderType=cv.BORDER_ISOLATED)
print("Second Convolved result: \n", result, '\n')

'''
Description: obtain the 1D Gaussian kernel using OpenCV's getGaussianKernel
@first argument: kernel size which should be odd
@second arguement: Gaussian standard deviation. If it is non-positive, it is computed from ksize as sigma = 0.3*((ksize-1)*0.5 - 1) + 0.8. 
'''
gaussian_1d_kernel = cv.getGaussianKernel(3, 2.7)
print("1D Gaussian Kernel:\n", gaussian_1d_kernel, "\n")

#obtain the 2D Gaussian kernel by multiplying the 1D kernel with its transpose
gaussian_2d_kernel = np.multiply(gaussian_1d_kernel, np.transpose(gaussian_1d_kernel))

#display the 2D Gaussian kernel
print("2D Gaussian Kernel:\n", gaussian_2d_kernel, "\n")

result = cv.filter2D(random_array, ddepth=-1, kernel=gaussian_2d_kernel, anchor=(-1, -1), borderType=cv.BORDER_ISOLATED)
print("2D Gaussian Convolved result: \n", result, '\n')
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#load the color image
unblurred_pattern = cv.imread("unblurred_pattern.tif", cv.IMREAD_COLOR)

#convert the image from BGR to RGB format using cv.cvtColor
unblurred_pattern_rgb = cv.cvtColor(unblurred_pattern, cv.COLOR_BGR2RGB) 

smoothed_pattern_rgb = cv.boxFilter(unblurred_pattern_rgb, -1, (21, 21))

#create a figure and plot the original and smoothed images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(unblurred_pattern_rgb)
ax[0].set_title("Original Image")
ax[0].axis("off")
ax[1].imshow(smoothed_pattern_rgb)
ax[1].set_title("Smoothed Image")
ax[1].axis("off")

#display the plot
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#load the color image
unblurred_pattern = cv.imread("unblurred_pattern.tif", cv.IMREAD_COLOR)

#convert the image from BGR to RGB format using cv.cvtColor
unblurred_pattern_rgb = cv.cvtColor(unblurred_pattern, cv.COLOR_BGR2RGB) 

#apply Gaussian blurring to smooth the image
smoothed_pattern_rgb = cv.GaussianBlur(unblurred_pattern_rgb, (21, 21), 20, 20)

#create a figure and plot the original and smoothed images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(unblurred_pattern_rgb)
axs[0].set_title("Original Image")
axs[0].axis("off")
axs[1].imshow(smoothed_pattern_rgb)
axs[1].set_title("Smoothed Image")
axs[1].axis("off")

#display the plot
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#load the color image
noisy_pattern = cv.imread("noisy_pattern.tif", cv.IMREAD_COLOR)

#convert the image from BGR to RGB format using cv.cvtColor
noisy_pattern_rgb = cv.cvtColor(noisy_pattern, cv.COLOR_BGR2RGB) 

#filter the noisy image using median blur
filtered_image_rgb = cv.medianBlur(noisy_pattern_rgb, 3)

#create a figure and plot the original and filtered images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(noisy_pattern_rgb)
axs[0].set_title("Original Image")
axs[0].axis("off")
axs[1].imshow(filtered_image_rgb)
axs[1].set_title("Filtered Image")
axs[1].axis("off")

#display the plot
plt.show()
```
